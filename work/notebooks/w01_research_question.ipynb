## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

### Decision
Help the SEO team decide which web pages should be refreshed first.

### Action
The SEO team reviews and updates the highest-priority pages before spending time on lower-priority pages.

### Unit of Analysis:
A single web page.


### Cost of a wrong recommendation
